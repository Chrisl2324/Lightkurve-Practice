```python
import lightkurve as lk
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# 1. Searching for a Light Curve File
- search by mission: Kepler, TESS, etc.
- search by Kepler quarter, TESS sector

Example using Kepler data:

```python
lc_search = lk.search_lightcurve('KIC 3733346', author='Kepler')
lc_search
```

Search by specific quarter:

```python
lc_search_quarter2 = lk.search_lightcurve('KIC 3733346', author='Kepler', quarter=2)
lc_search_quarter2
```

# 2. Downloading Light Curve Data
- download single file or collection of multiple light curves

Single Light Curve:

```python
lc = lc_search_quarter2.download()
lc
```

Plot data using matplotlib plot method:

```python
lc.plot()
```

Mupltiple Light Curves:

```python
lc_collection = lc_search[:3].download_all()
lc_collection
```

Plotting multiple light curves:

```python
lc_collection.plot()
```

Normalize data and scale the axes:

```python
fig, ax = plt.subplots(figsize=(20, 5))
for lc in lc_collection:
    lc.normalize().plot(ax=ax, label=f'Quarter {lc.quarter}');
```

# 3. Searching Target Pixel Files

Using K2-199:

```python
tpf_search = lk.search_targetpixelfile('K2-199', exptime=1800)
tpf_search
```

Searching for specific tpf:

# 4. Downloading Target Pixel File and Plotting:
- single or collection of files

Single tpf:

```python
tpf = tpf_search.download()
tpf.plot()
```

Accessing data values:

```python
tpf.flux[0]
```

Collection of tpf:

```python
tpf_collection = tpf_search.download_all()
tpf_collection.plot()
```

Using Aperture Mask:

```python
tpf.pipeline_mask
tpf.plot(aperture_mask=tpf.pipeline_mask)
```

Inspecting Raw Data from file:

```python
tpf.get_header()
```

Coverting data to light curve file:

```python
lc = tpf.to_lightcurve()
lc.plot()
```

# 6. Removing Noise from Light Curve using PLDCorrector

```python
from lightkurve.correctors import PLDCorrector
pld = PLDCorrector(tpf)
print(pld)
```

```python
corrected_lc = pld.correct()
```

Plotting Uncorrected Light Curve vs Corrected 

```python
original_lc = lc
ax = original_lc.normalize().scatter(color='blue', label='Uncorrected LC')
corrected_lc.normalize().remove_outliers().scatter(ax=ax, color='black', label='Corrected LC')
```

Diagnose success of correction:

```python
pld.diagnose()
```

```python
pld.diagnose_masks()
```
